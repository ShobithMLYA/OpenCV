```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
full_image = cv2.imread("assets//FullBody.jpg")
plt.imshow(full_image)
plt.show()
```

```python
fullbody_cascade = cv2.CascadeClassifier('Haarcascades/haarcascade_fullbody.xml')
```

```python
gray_image = cv2.cvtColor(full_image, cv2.COLOR_BGR2GRAY)
body = fullbody_cascade.detectMultiScale(gray_image, 1.03,)
result = full_image.copy()
for (bx,by,bw,bh) in body:
    cv2.rectangle(result,(bx,by),((bx + bw),(by + bh)),(255,0,0),2)
```

```python
plt.imshow(result)
plt.show()
```
