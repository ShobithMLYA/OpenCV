```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.ones((500, 500, 3), dtype = 'uint8')*255
a = image.copy()
%matplotlib inline
plt.imshow(a)
plt.show()
```

```python
line_1 = cv2.line(a, (10,10), (490,490), (255, 0, 0), 3)
line_2 = cv2.line(a, (10,490), (490,10), (255, 0, 0), 3)
plt.imshow(a)
plt.show()
```

```python
b = image.copy()
rect = cv2.rectangle(b, (0,0), (500,500), (0, 255, 0), 10)
plt.imshow(b)
plt.show()
```

```python
c = image.copy()
(x, y) = int(image.shape[1]/2), int(image.shape[0]/2)
circle = cv2.circle(c, (x, y), (200), (0, 0, 255), 3)
plt.imshow(c)
plt.show()
```
