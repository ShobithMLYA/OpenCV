```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.ones((400, 400, 3), dtype = 'uint8')*255
a = image.copy()
%matplotlib inline
plt.imshow(a)
plt.show()
```

```python
b = image.copy()
(x, y) = int(image.shape[1]/2), int(image.shape[0]/2)
ellipse = cv2.ellipse(b, (x, y), (100, 25), 0, 0, 360, (0, 255, 0), 3)
plt.imshow(b)
plt.show()
```

```python
c = image.copy()
quad = np.array([[0,200], [200, 0], [400, 200], [200, 400]], np.int64)
poly = cv2.polylines(c, [quad], True, (255, 0, 0), 3)
plt.imshow(poly)
plt.show()
```
