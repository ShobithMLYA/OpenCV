```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
smile_image = cv2.imread('assets/smile.jpg')
plt.imshow(smile_image)
plt.show()
```

```python
smile_cascade = cv2.CascadeClassifier('Haarcascades/haarcascade_smile.xml')
```

```python
gray_image = cv2.cvtColor(smile_image, cv2.COLOR_BGR2GRAY)
body = smile_cascade.detectMultiScale(gray_image, 3,3)
result = smile_image.copy()
for (bx,by,bw,bh) in body:
    cv2.rectangle(result,(bx,by),((bx + bw),(by + bh)),(255,0,0),2)
```

```python
plt.imshow(result)
plt.show()
```
