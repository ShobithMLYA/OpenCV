```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((400, 400, 3), dtype = 'uint8')

a=image.copy()
rectangle = cv2.rectangle(a, (50,50), (350,350), (255, 255,255 ), -1)
plt.imshow(rectangle)
plt.show()

b = image.copy()
circle = cv2.circle(b, (200, 200), (150),(255,255,255), -1)
plt.imshow(circle)
plt.show()
```

```python
c = image.copy()
bitwise_XOR = cv2.bitwise_xor(rectangle, circle)
plt.imshow(bitwise_XOR)
plt.show()
```

```python
d = image.copy()
bitwise_NOT = cv2.bitwise_not(rectangle, circle)
plt.imshow(bitwise_NOT)
plt.show()
```
